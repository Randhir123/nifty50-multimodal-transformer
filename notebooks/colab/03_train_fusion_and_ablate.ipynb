# 03 — Train fusion variants and run ablations

This notebook compares modality combinations using the same aligned artifact.

Variants:

- `tabular_only`
- `tabular_kg`
- `tabular_image`
- `tabular_text`
- `tabular_image_text_kg`

These are classification/ablation metrics, not yet a portfolio backtest.

````python
import os
REPO_DIR = "/content/nifty50-multimodal-transformer"
if os.path.exists(REPO_DIR):
    os.chdir(REPO_DIR)
print("Working directory:", os.getcwd())
````

````python
from pathlib import Path
import subprocess, sys
import pandas as pd

OUTPUT_DIR = Path("data/processed/real_world_demo")
DATASET = OUTPUT_DIR / "real_world_multimodal_samples.npz"
ABLATION_DIR = OUTPUT_DIR / "ablations"

cmd = [
    sys.executable,
    "scripts/run_ablation_study.py",
    "--dataset", str(DATASET),
    "--output-dir", str(ABLATION_DIR),
    "--epochs", "1",
    "--batch-size", "4",
    "--device", "cpu",
    "--model-dim", "16",
    "--num-heads", "4",
    "--num-layers", "1",
    "--ff-dim", "32",
    "--val-fraction", "0.25",
]
print(" ".join(cmd))
subprocess.run(cmd, check=True)
````

````python
results = pd.read_csv(ABLATION_DIR / "ablation_results.csv")
display(results)
````

````python
# Plot ablation metrics from actual run outputs.
import matplotlib.pyplot as plt

metrics = ["accuracy", "precision", "recall", "f1", "roc_auc"]
plot_df = results.set_index("variant")[metrics]

ax = plot_df.plot(kind="bar", figsize=(12, 5))
ax.set_title("Ablation metrics from real-world demo artifact")
ax.set_ylabel("Metric")
ax.set_ylim(0, 1)
plt.xticks(rotation=30, ha="right")
plt.tight_layout()

output_path = OUTPUT_DIR / "ablation_metrics_bar_chart.png"
plt.savefig(output_path, dpi=160)
print("Wrote:", output_path)
plt.show()
````
